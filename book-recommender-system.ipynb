## Popularity based recommender system

### we are going to display highest average rating books on which, minimum 250 people have rated.

#### Step 1 :
##### we are merging books and rating column under ISBN 

```python
ratings.merge(books, on ='ISBN')
```

```python
ratings.merge(books, on ='ISBN').shape
```

```python
ratings_with_bookName = ratings.merge(books, on ='ISBN')
```

```python
ratings_with_bookName
```

#### Step 2: groupby on book title basis

```python
ratings_with_bookName.groupby('Book-Title').count()
```

```python
ratings_with_bookName.groupby('Book-Title').count()['Book-Rating'] 
```

```python
ratings_with_bookName.groupby('Book-Title').count()['Book-Rating'].reset_index()
```

```python
num_rating_df = ratings_with_bookName.groupby('Book-Title').count()['Book-Rating'].reset_index()
```

```python
num_rating_df.rename(columns={'Book-Rating':'num_of_ratings'}, inplace=True)
```

```python
num_rating_df
```

```python
avg_rating_df = ratings_with_bookName.groupby('Book-Title', as_index=False)['Book-Rating'].mean()
```

```python
avg_rating_df.rename(columns={'Book-Rating':'avg_ratings'},inplace=True)
```

```python
avg_rating_df
```

```python
popularity_df = num_rating_df.merge(avg_rating_df, on = 'Book-Title')
```

```python
popularity_df
```

```python
popularity_df[popularity_df['num_of_ratings']>=50]
```

```python
popularity_df[popularity_df['num_of_ratings']>=50].sort_values('avg_ratings', ascending=False)
```

```python
popularity_df[popularity_df['num_of_ratings']>=250].sort_values('avg_ratings', ascending=False).head(50)
```

```python
popularity_df = popularity_df[popularity_df['num_of_ratings']>=250].sort_values('avg_ratings', ascending=False).head(50)
```
