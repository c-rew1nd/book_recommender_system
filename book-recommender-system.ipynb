#### Starting colab filter based recommender system

```python
ratings_with_bookName
```

```python
ratings_with_bookName.groupby('User-ID').count()['Book-Rating']
```

```python
x = ratings_with_bookName.groupby('User-ID').count()['Book-Rating'] > 199
```

```python
x[x] #boolean indexing
```

```python
uwm2r = x[x].index # uwm2r = users who have done minimum 200 useres 
```

```python
uwm2r
```

```python
ratings_with_bookName['User-ID'].isin(uwm2r)
```

```python
ratings_with_bookName[ratings_with_bookName['User-ID'].isin(uwm2r)]
```

```python
filtered_rating = ratings_with_bookName[ratings_with_bookName['User-ID'].isin(uwm2r)]
```

```python
filtered_rating.groupby('Book-Title').count()['Book-Rating']
```

```python
y= filtered_rating.groupby('Book-Title').count()['Book-Rating'] > 50
```

```python
y[y]
```

```python
famous_books = y[y].index
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
final_ratings
```

```python
final_ratings.drop_duplicates()
```

```python
final_ratings.drop_duplicates('Book-Title')
```

```python
pt = final_ratings.pivot_table(index='Book-Title',columns='User-ID', values='Book-Rating')
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pt)
```

```python
similarity_scores
```

```python
sorted(list(enumerate(similarity_scores[0])),key=lambda x:x[1],reverse=True)
```

```python
def recommend(book_name):
    index = np.where(pt.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:11]
    for i in similar_items:
        print(pt.index[i[0]])
```

```python
recommend('Message in a Bottle')
```
